#**Step 07: Create a Large Language Model (LLM) Wrapper**

```python
#llm=ChatOpenAI()
```

```python
#llm
```

```python
notebook_login()
```

```python
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-chat-hf",
                                          use_auth_token=True,)


model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-2-7b-chat-hf",
                                             device_map='auto',
                                             torch_dtype=torch.float16,
                                             use_auth_token=True,
                                              load_in_8bit=True,
                                              #load_in_4bit=True
                                             )
```

```python
pipe = pipeline("text-generation",
                model=model,
                tokenizer= tokenizer,
                torch_dtype=torch.bfloat16,
                device_map="auto",
                max_new_tokens = 512,
                do_sample=True,
                top_k=30,
                num_return_sequences=1,
                eos_token_id=tokenizer.eos_token_id
                )
```

```python
llm=HuggingFacePipeline(pipeline=pipe, model_kwargs={'temperature':0})
```

```python
llm.predict("Please provide a concise summary of the article Navegamos hacia la segunda fase de la IA generativa. La frontera de las realidades")
```

#**Step 08: Initialize the Retrieval QA with Source Chain**

```python
chain = RetrievalQAWithSourcesChain.from_llm(llm=llm, retriever=vectorstore.as_retriever())
```

```python
result=chain({"question": "cual es el mayor reto de esta tecnología?"}, return_only_outputs=True)
```
